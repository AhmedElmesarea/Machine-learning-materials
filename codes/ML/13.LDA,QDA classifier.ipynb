```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
from sklearn.datasets import load_iris,load_breast_cancer

from sklearn.metrics import  confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import warnings 
warnings.filterwarnings('ignore')
```

```python
iris=load_iris()
```

```python
x=iris.data
y=iris.target
```

```python
x.shape
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
scale=StandardScaler()
X=scale.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=10)
```

```python
model=LinearDiscriminantAnalysis(n_components=2)
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
m=LogisticRegression()
m.fit(x_train,y_train)
```

```python
m.score(x_train,y_train)
```

```python
m.score(x_test,y_test)
```

```python
cancer=load_breast_cancer()
```

```python
input=cancer.data
label=cancer.target
```

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis,LinearDiscriminantAnalysis
```

```python
input.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(input,label,test_size=0.2,random_state=10)
```

```python
LR=LogisticRegression()
LDA=LinearDiscriminantAnalysis()
QDA=QuadraticDiscriminantAnalysis()
```

```python
LR.fit(x_train,y_train)
```

```python
LDA.fit(x_train,y_train)
```

```python
QDA.fit(x_train,y_train)
```

```python
print(f'accuracy trainnig in LR : {LR.score(x_train,y_train)}')
print(f'accuracy testing in LR : {LR.score(x_test,y_test)}')
```

```python
print(f'accuracy trainnig in LDA : {LDA.score(x_train,y_train)}')
print(f'accuracy testing in LDA : {LDA.score(x_test,y_test)}')
```

```python
print(f'accuracy trainnig in QDA: {QDA.score(x_train,y_train)}')
print(f'accuracy trainnig in QDA: {QDA.score(x_test,y_test)}')
```
