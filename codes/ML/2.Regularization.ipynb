#### Regularization to prevent overfitting

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import warnings
warnings.filterwarnings('ignore')
```

```python
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]
df = pd.DataFrame({'x': x, 'y': y})
df
```

```python
sns.scatterplot(x=df['x'],y=df['y'])
plt.title('Data')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
x=df.drop('y',axis=1)
Y=df['y']
```

```python
model.fit(x,y)
```

```python
sns.scatterplot(x=df['x'],y=df['y'],label='real values')
plt.plot(x,model.predict(x),color='r',label='prediction')
plt.legend()
plt.title('let is overfitting')
plt.show()
```

Regularization types:
- Ridge -->reduce parameters of features (dont make it =0)
- Lasso -->reduce parameters of features (maybe make it =0)

```python
from sklearn.linear_model import Ridge,Lasso
```

```python
ridge=Ridge(alpha=0.4)
lasso=Lasso(alpha=0.4)
```

```python
ridge.fit(x,y)
```

```python
sns.scatterplot(x=df['x'],y=df['y'],label='real values')
plt.plot(x,ridge.predict(x),color='r',label='prediction')
plt.legend()
plt.title('prevent overfitting using Ridge')
plt.show()
```

```python
lasso.fit(x,y)
```

```python
sns.scatterplot(x=df['x'],y=df['y'],label='real values')
plt.plot(x,lasso.predict(x),color='r',label='prediction')
plt.legend()
plt.title('prevent overfitting using lasso')
plt.show()
```

```python
plt.figure(figsize=(14,9))
plt.subplot(2,2,1)
sns.scatterplot(x=df['x'],y=df['y'],label='real values')
plt.plot(x,model.predict(x),color='r',label='prediction')
plt.legend()
plt.title('linear regression')


plt.subplot(2,2,2)
sns.scatterplot(x=df['x'],y=df['y'],label='real values')
plt.plot(x,ridge.predict(x),color='r',label='prediction')
plt.legend()
plt.title(' Ridge')

plt.subplot(2,2,3)
sns.scatterplot(x=df['x'],y=df['y'],label='real values')
plt.plot(x,lasso.predict(x),color='r',label='prediction')
plt.legend()
plt.title('lasso')
plt.show()



plt. show()
```
